### 1. Rock, Paper , Scissors

```python
import random

def computer_pick_for_human_input(predicted_user_input):
    '''Computer pick based on the predicted user input - function called in the computer_final_selection function below'''
    if predicted_user_input == 'rock':
        computer_move = 'paper'
    elif predicted_user_input == 'paper':
        computer_move = 'scissors'
    elif predicted_user_input == 'scissors':
        computer_move = 'rock'
    return computer_move

def computer_final_selection(user_inputs_all):
    '''Output computer's move on the basis of the all the human's inputs till now'''
    user_max_played_option = max(user_inputs_all.values())
    max_key=[]

    #get the options which the user has played the most
    for key, value in user_inputs_all.items():
        if value == user_max_played_option:
            max_key.append(key)
            
    #if the user has played three options equal number of times, computer makes a random choice between these three 
    #but if two options are played equally, computer brain plays that option with which it is more likely to win
    if len(max_key)==1:
        predicted_user_input = max_key[0]
        computer_move = computer_pick_for_human_input(predicted_user_input)
    elif len(max_key)==2:
        if sorted(max_key)== ['paper', 'rock']:
            computer_move = 'paper'
        elif sorted(max_key)== ['paper', 'scissors']:
            computer_move = 'scissors'
        elif sorted(max_key)== ['rock', 'scissors']:
            computer_move = 'rock'
    elif len(max_key)==3:
        predicted_user_input = random.choice(max_key)
        computer_move = computer_pick_for_human_input(predicted_user_input)
    
    return computer_move
```

```python
#a complete log of what the user has played uptil now
user_inputs_all={'rock':0, 'paper':0, 'scissors':0}

games_played=0
user_wins=0
user_losses=0
comp_user_ties=0


validity_flag=1

while validity_flag==1:
    
    computer_plays = computer_final_selection(user_inputs_all)
    
    user_input = input('Please select a valid choice among Rock/Paper/Scissors/Quit:')
    user_input= user_input.lower()
    
    while user_input not in ['rock', 'paper', 'scissors', 'quit']:
        user_input = input('Invalid Input! \nPlease select a valid choice among Rock/Paper/Scissors/Quit:')
        user_input= user_input.lower()
    
    if user_input== 'rock':
        user_inputs_all['rock']+=1
    elif user_input == 'scissors':
        user_inputs_all['scissors']+=1
    elif user_input== 'paper':
        user_inputs_all['paper']+=1
    
    
    if computer_plays== user_input:
        comp_user_ties+=1
        games_played+=1
        print('You both played {}, its a tie\n'.format(computer_plays))
        
    elif user_input =='rock':
        if computer_plays=='paper':
            user_losses+=1
            games_played+=1
            print('Computer played : {}, You played: {}, You lose!\n'.format(computer_plays,user_input))
        elif computer_plays=='scissors':
            user_wins+=1
            games_played+=1
            print('Computer played : {}, You played: {}, You win!\n'.format(computer_plays,user_input))
            
    elif user_input =='paper':
        if computer_plays=='rock':
            user_wins+=1
            games_played+=1
            print('Computer played : {}, You played: {}, You win!\n'.format(computer_plays,user_input))
        elif computer_plays=='scissors':
            user_losses+=1
            games_played+=1
            print('Computer played : {}, You played: {}, You lose!\n'.format(computer_plays,user_input))
            
    elif user_input =='scissors':
        if computer_plays=='rock':
            user_losses+=1
            games_played+=1
            print('Computer played : {}, You played: {}, You lose!\n'.format(computer_plays,user_input))
        elif computer_plays=='paper':
            user_wins+=1
            games_played+=1
            print('Computer played : {}, You played: {}, You win!\n'.format(computer_plays,user_input))
            
    elif user_input =='quit':
        print('User chose to Quit\n')
        print('Games Played:{}, Matches won by human: {}'.format(games_played,user_wins))
        validity_flag=0
```

### 3. Near Duplicate detection

#### 3(a)

```python
import re 
import math

#open and read the file
fp = open("Santa.txt", 'r', encoding="ISO-8859-1")
lines=fp.readlines()


def moving_window(single_tweet):
    ''' the function takes as input a tweet (that is,a string), and converts it into a bunch of 3-word phrases'''
    word_list_in_tweet= single_tweet.strip().split(' ')
    
    phrases_dict={}
    
    for i in range(0,len(word_list_in_tweet)-2):
        phrase= ' '.join(word_list_in_tweet[i:i+3])
        phrases_dict[phrase]=1
    
    return phrases_dict
    

#print the output for all tweets
for line in lines:
    phrases_dict= moving_window(line)
    print(phrases_dict)
```

#### 3(b)

```python
def cosine(tweet1_phrases_dict, tweet2_phrases_dict):
    '''This functions computes the cosine similairty between two tweets -takes dictionary of phrases for each tweet as the two inputs'''
    matches = 0
    
    #compute number of matches between the 2 tweets
    for tweet1_phrase in tweet1_phrases_dict.keys():
        for tweet2_phrase in tweet2_phrases_dict.keys():
            if tweet1_phrase.lower() == tweet2_phrase.lower():
                matches+=1
                
    cosine_similarity= matches/math.sqrt(len(tweet1_phrases_dict.keys()) * len(tweet2_phrases_dict.keys()))
    
    return cosine_similarity
```

```python
phrases_dict_list = []

#the following loops are for generating a sample output of cosine similarity between two tweets
for line in lines:
    phrases_dict= moving_window(line)
    phrases_dict_list.append(phrases_dict)

for i in range(3,len(phrases_dict_list)-1):
    print('Phrases for TWeet 1:', phrases_dict_list[i],'\n')
    print('Phrases for Tweet 2',phrases_dict_list[i+1], '\n')
    cosine_sim_val = cosine(phrases_dict_list[i], phrases_dict_list[i+1])
    print('Cosine Similarity',cosine_sim_val,'\n')
```

#### 3(C)

```python
def near_duplicates(tweets):
    ''' This function reads in the tweets from Santa.txt and outputs the tweet and a near duplicate of that tweet among the previously seen tweets'''
    for i in range(1,len(tweets)):
        tweet1= tweets[i]
        for j in range(i,0,-1):
            tweet2 = tweets[j-1]
            
            tweet1_phrases_dict= moving_window(tweet1)
            tweet2_phrases_dict= moving_window(tweet2)
            
            #find cosine similarity between the two tweets
            cosine_sim_val = cosine(tweet1_phrases_dict, tweet2_phrases_dict)
            
            #is cosine similarity is greater than 0.5, then print the two tweets
            if cosine_sim_val > 0.5:
                print('Near duplicate tweets: \n1.{}\n2.{}\n'.format(tweet2,tweet1))
                
#print the output
near_duplicates(lines)
```

### 4. The Google of Quotes

#### 4(a) 
Build a list of full quotes (5 points). Read in the file, and create a list of full quotes of the form “quote - speaker”. For example, “The heart has its reasons, of which the mind knows nothing. - Blaise Pascal”.

```python
#open the file and read it
fp = open('quotes.txt', 'r', encoding="ISO-8859-1")
lines = fp.readlines()

quotes_list= []

#iterate over quote and author in every line and append the two
for i in range(0,len(lines),2):
        quote_with_author = lines[i].rstrip('\n')+' - '+ lines[i+1].rstrip('\n')
        quotes_list.append(quote_with_author)

fp.close()

#print the ouput
quotes_list
```

#### 4(b)

```python
import re

word_list=[]

def get_words_from_quotes(quote):
    '''This function takes a quote as an input and outputs a list of the words in the it-  all are lower-case, and  contains only characters, digits, or
underscore'''
    return re.split('[^a-zA-Z0-9_]+', quote.strip('').lower())
    
for i in quotes_list:
    word_list.append(get_words_from_quotes(i))

print(word_list)
```

#### 4(c)

```python
def build_pl_dict(quotes_list):

    pl_dict = {}

    for i in quotes_list:
        words_in_quotes = get_words_from_quotes(i)

        word_dict = {}

        for word in words_in_quotes:
            if word in word_dict:
                word_dict[word]+=1
            else:
                word_dict[word]=1

        pl_dict[i]= word_dict

    return pl_dict

#print output
pl_dict= build_pl_dict(quotes_list)
print(pl_dict)
```

```python
def build_reverse_pl_dict(quotes_list): 
    '''We build the reverse postings-list in this function where the output is a dictionary whose keys are the words, and the values are \
themselves dictionaries with the key being a full quote, and the value being the number of times the word appeared in the full quote'''
    reverse_pl_dict = {}

    unique_words_list= []

    for quote in quotes_list:
        words_in_quotes = get_words_from_quotes(quote)
        for word in words_in_quotes:
            if word not in unique_words_list:
                unique_words_list.append(word)

    pl_dict = build_pl_dict(quotes_list)

    for unique_word in unique_words_list:
        reverse_pl_dict_values = {}
        for pl_dict_key, pl_dict_value in pl_dict.items():
            if unique_word in pl_dict_value.keys():
                reverse_pl_dict_values[pl_dict_key]= pl_dict_value[unique_word]
        
        reverse_pl_dict[unique_word]=reverse_pl_dict_values
    return reverse_pl_dict
```

```python
reverse_pl_dict = build_reverse_pl_dict(quotes_list)

print(reverse_pl_dict['how'])
```

#### 4(E)

```python
import math

def tf_idf(word, quote, quotes_list):
    ''' This function computes the TF-IDF measure for every word in the provided quote '''
    pl_dict = build_pl_dict(quotes_list)
    
    #compute TF
    for pl_dict_key, pl_dict_value in pl_dict.items():
        if pl_dict_key==quote:
            for pl_dict_value_key, pl_dict_value_value in pl_dict_value.items():
                if pl_dict_value_key == word:
                    all_values = pl_dict_value.values()
                    tf_den = max(all_values)
                    tf_num = int(pl_dict_value_value)
                    tf = tf_num/tf_den
                    
    total_num_of_quotes = len(quotes_list)
    
    #computer IDF
    reverse_pl_dict = build_reverse_pl_dict(quotes_list)
    for reverse_pl_dict_key, reverse_pl_dict_value in reverse_pl_dict.items():
        if reverse_pl_dict_key==word:
            ratio = total_num_of_quotes/len(reverse_pl_dict_value.keys())
            idf = math.log(ratio)

    
    #calculate TF IDF score
    tf_idf_score = tf*idf
    return tf_idf_score
```

#### 4(f)

```python
def word_tf_idf(word):
    '''This functions takes a word as argument, and returns a dictionary whose keys are full quotes containing that word, 
    and whose values are the TF-IDF score of that word for that full quote '''
    output_dict= {}
    reverse_pl_dict = build_reverse_pl_dict(quotes_list)
    
    for reverse_pl_dict_key,reverse_pl_dict_value in reverse_pl_dict.items():
        if reverse_pl_dict_key==word.lower():
            for reverse_pl_dict_value_key, reverse_pl_dict_value_value in reverse_pl_dict_value.items():
                tf_idf_measure= tf_idf(word, reverse_pl_dict_value_key, quotes_list)
                output_dict[reverse_pl_dict_value_key]=tf_idf_measure

    return output_dict
```

```python
def multiple_word_search(search_words_list):
    ''' This function takes a list of words as argument, and returns a dictionary whose keys are full quotes containing one or 
    more of the words in that list, and whose values are the sum of TF-IDF scores of the words in that list for that full
    quote '''
    
    first_word_tf_idf_dict = word_tf_idf(search_words_list[0])
    search_words_list.pop(0)

    for search_word in search_words_list:
        temp_if_idf_dict = word_tf_idf(search_word)
        for temp_if_idf_dict_key, temp_if_idf_dict_value in temp_if_idf_dict.items():
                if temp_if_idf_dict_key in first_word_tf_idf_dict.keys():
                    first_word_tf_idf_dict[temp_if_idf_dict_key]  += temp_if_idf_dict_value
                else:
                    first_word_tf_idf_dict[temp_if_idf_dict_key]=temp_if_idf_dict[temp_if_idf_dict_key]
                    
    return first_word_tf_idf_dict
```
